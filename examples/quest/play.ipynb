```python
import pickle
```

```python
with open("data/raw_data_qasm/quest_rbf.data", "rb") as f:
    data = pickle.load(f)
```

```python
import qiskit
def num_used_qubits(circuit):
    return len(set([q for instr in circuit.data for q in instr.qubits]))
```

```python
cleaned_data = []
for circ_qasm, props, label in data:
    circ = qiskit.QuantumCircuit.from_qasm_str(circ_qasm)
    if num_used_qubits(circ) > 0:
        cleaned_data.append((circ_qasm, props, label))
```

```python
with open("data/raw_data_qasm/quest_rbf.data", "wb") as f:
    pickle.dump(cleaned_data, f)
```
